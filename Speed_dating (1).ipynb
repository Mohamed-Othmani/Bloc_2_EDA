```python
#data_unique_woman = data_unique.loc[data_unique["Genre"] == "woman"]
#data_unique_man = data_unique.loc[data_unique["Genre"] == "man"]




#data_unique = data.groupby("iid").first().reset_index()
#data_unique 
#data_unique["Genre"].value_counts()

data_unique_woman = data_unique[data_unique["Genre"] == "Woman"]
data_unique_man = data_unique[data_unique["Genre"] == "Man"]

# What women think men are looking for, compared to what men are really looking for.

# Compute means for each attribute for women and men separately
mean_attributes_women = data_unique_woman[["attr2_1", "sinc2_1", "intel2_1", "fun2_1", "amb2_1", "shar2_1"]].mean()
mean_attributes_men = data_unique_man[["attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"]].mean()

categories = ['attirante', 'sincère', 'intelligente', 'amusante', 'ambitieuse', 'centres d\'intérêt partagés']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=mean_attributes_men,
    theta=categories,
    fill='toself',
    name='Ce que les hommes cherchent'
))

fig.add_trace(go.Scatterpolar(
    r=mean_attributes_women,
    theta=categories,
    fill='toself',
    name='Ce que les femmes pensent que les hommes cherchent'
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 40])),
    showlegend=True,
    title_text="Ce que les femmes pensent que les hommes cherchent VS ce que les hommes cherchent",
    title_x=0.42
)

fig.show()
```

on vois que les femmes penses sur estimes les attentes des hommes en termes d'attirance et sous estimes leur attentes intelectuels et d'honneteté

```python
mean_att_others_by_men = data_unique_man["attr2_1"].mean()
mean_sinc_others_by_men = data_unique_man["sinc2_1"].mean()
mean_inte_others_by_men = data_unique_man["intel2_1"].mean()
mean_fun_others_by_men = data_unique_man["fun2_1"].mean()
mean_ambi_others_by_men = data_unique_man["amb2_1"].mean()
mean_shar_others_by_men = data_unique_man["shar2_1"].mean()

mean_att_women = data_unique_woman["attr1_1"].mean()
mean_sinc_women = data_unique_woman["sinc1_1"].mean()
mean_inte_women = data_unique_woman["intel1_1"].mean()
mean_fun_women = data_unique_woman["fun1_1"].mean()
mean_ambi_women = data_unique_woman["amb1_1"].mean()
mean_shar_women = data_unique_woman["shar1_1"].mean()

categories = ['attirante', 'sincère', 'intelligente', 'amusante', 'ambitieuse', 'centres d\'intérêt partagés']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
                  r=[mean_att_others_by_men,mean_sinc_others_by_men, mean_inte_others_by_men, mean_fun_others_by_men,
                  mean_ambi_others_by_men,mean_shar_others_by_men],
                  theta=categories,
                  fill='toself',
                  name='Ce que les hommes pensent que les femmes cherchent'
))

fig.add_trace(go.Scatterpolar(
                  r=[mean_att_women,mean_sinc_women, mean_inte_women, mean_fun_women, mean_ambi_women, mean_shar_women],
                  theta=categories,
                  fill='toself',
                  name='Ce que les femmes cherchent '
))

fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0, 40])),showlegend=True,
                  title_text="Ce que les hommes pensent que les femmes cherchent VS ce que les femmes cherchent", title_x=0.42)

fig.show()
```

```python
# Récupération des données
decisions = data_unique["dec"].value_counts()


# Paramètres pour la figure
fig, ax = plt.subplots(figsize=(8, 8))
colors = ['#66b3ff', '#ff9999']
labels = decisions.index.tolist()
explode = [0.05, 0]

# Création du camembert
wedges, texts, autotexts = ax.pie(decisions, colors=colors, labels=labels, explode=explode,
                                   autopct='%1.1f%%', startangle=90, pctdistance=0.75,
                                   textprops=dict(color="w"))

# Ajout d'un cercle au centre
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig.gca().add_artist(centre_circle)

# Réglage des paramètres du texte
for text, autotext in zip(texts, autotexts):
    text.set_color('black')
    text.set_fontsize(14)
    autotext.set_color('black')
    autotext.set_fontsize(14)

# Réglage des paramètres de la légende
ax.legend(wedges, labels, title="Décisions", loc="center", bbox_to_anchor=(0.5, -0.15), fontsize=14, title_fontsize=16)

# Ajout d'un titre
plt.title("Répartition des décisions en pourcentage", fontsize=18)

plt.show()
```
